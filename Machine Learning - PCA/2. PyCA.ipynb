```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set();
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data.data, columns= data.feature_names)
df
```

## Here, we need to know `Which features are the most important ones that explain the variance in the data`...

Step 1 - PCA  
Step 2 - ML Algorithm

## 

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
scaled_df = pd.DataFrame(std.fit_transform(df))
```

#### Trick to remove the 0.0001200E10 kind of values ↓

```python
scaled_df.describe().apply(lambda x: round(x, 3))
```

# 

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components= 2)
```

```python
pca_ed = pca.fit_transform(scaled_df)
pca_ed
```

```python
pca_ed.shape
```

```python
plt.style.use('dark_background')
plt.figure(figsize= (20, 10))
plt.scatter(pca_ed[:, 0], pca_ed[:, 1], color= 'red')
```

### Interpreting these components is not easy to understand...

```python
plt.style.use('dark_background')
plt.figure(figsize= (20, 10))
plt.scatter(pca_ed[:, 0], pca_ed[:, 1], c= data['target'], cmap= 'rainbow')
```

```python
pca.components_
```

```python
comps = pd.DataFrame(pca.components_, columns= data.feature_names)
comps
```

```python
plt.figure(figsize= (20, 10))
sns.heatmap(comps, cmap= 'Blues')
```
