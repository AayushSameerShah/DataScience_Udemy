# Style and Fonts for Seaborn Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

### `sns.set_style` to do major stuff

```python
# Without
sns.countplot(x= "sex", data= tips)
```

```python
# With
sns.set_style('dark')                                                # Choose from: white, dark, whitegrid, darkgrid, ticks
sns.countplot(x='sex', data=tips)
```

#### Making plot more 'Neat'
Where the LINE is named as **Spine** so `de-`**`spine`** !

```python
sns.countplot(x='sex', data=tips)
sns.despine()
```

#### More de-spining...

```python
sns.countplot(x='sex', data=tips)
sns.despine(left= True, bottom= True)
```

## Changing the Size of plot... (different than `aspect` & `size`)

It is where we added the `fig = plt.figure((size,size))` and then <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;`axes = fig.add_axes([1,1,1,1])`

```python
fig = plt.figure(figsize= (10,10))
sns.countplot(x='sex', data= tips)
```

## Wah!
But I still want `fonts` more sized!!


```python
plt.figure(figsize= (10,10))
sns.set_context('poster')                                         # paper, notebook, talk, poster
sns.countplot(x= 'sex', data= tips)
```

#### Works, 
Now, in our size! ( I want even more! )

```python
plt.figure(figsize= (10,10))
sns.set_context('poster', font_scale= 2)
sns.countplot(x= 'sex', data= tips)
```
