# Grid Plotting
Up untill now, we have seen various kinds of plotting types, but here we will see the new **technique** to plot the plot!

```python
import seaborn as sns
```

```python
flowers = sns.load_dataset('iris')
flowers.head()
```

### Now, 
Whenever we get the dataset, it is a good practice to plot the data in front of your face to 'see' what's going on here.
<br>
The common way is to use `sns.pairplot()`, that's it.

```python
sns.pairplot(flowers)
```

### Yes, it has given the data, BUT,

    Agreed, using this technique is relatively easy, as it does everything for you in just 
    one line, but it doesn't give you `control` to do stuff.
    
Let's gain more control over how the data should be presented and which plotting type should be use with **Gridplots**.

```python
sns.PairGrid(flowers)
```

### Wov, Wov, Wov,
It gave just an empty set of plots!
<br> *Obviously* man!
<br> We have to give it how these data should be plotted.

```python
empty_plots = sns.PairGrid(flowers)

#Filling it
empty_plots.map(plt.scatter)
```

#### Huli shuli... Bulli... ***
Whole stuff has been filled with scatter only!
Ok, let me try giving more of them like we did for the `scatter`... I mean multiple times...

```python
empt = sns.PairGrid(flowers)
empt.map(plt.scatter)             # Applied
empt.map(sns.distplot)            # Not applied
```

### Looks like `.map()` is not workning... It only works for the plt plots. (Yes you can write 2 map() and pass plt plots and will work)

Now to plot with sns plots, we will use its (PairGrid's) own `map functions`.

They are:

     .map_diag                  For Normal diagonal
     .map_lower                 For Lower left  Half
     .map_offdiag                         -
     .map_upper                 For Upper right Half


```python
empt = sns.PairGrid(flowers)
empt.map_upper(plt.scatter)
empt.map_diag(sns.histplot)
empt.map_lower(sns.kdeplot)
```

### Cool. 
Don't be overwhelmed to this. The PairGrid only gives you the control over what kind of plot you want to plot like.
<br> (Yes, there are the parameters to tweak the result, but for now to discuss the following topic, let's skip that...)

### So, what is Facet Grid?
    
    1.

    (  )

#### So simple definition and explaination... just read... again READ!

That means, (please now forget about those PairGrid and pairplot - they are different) when you want to see what is the result when its and it is not. 

In one word: <b> <font size=3 face=nexa> It is your type of thing! </font></b>

    Here, how you want to see in the, let's say there are three things A, B, C. And they have their two stats
    On / Off.
    
    Based on their stats, a program gives different result.
    
    You see?
    You want to see the results in all conditions, like what if A is ON and B is OFF and C is ON?
                                                        what if A is OFF and B is OFF and C is ON?
                                                        
                                                        And all the combinations...
                                                        3n3...
                                                        
                                                        So... This is the same thing.
                                                        Facet = All combinations.
                                                        
#### What is the advantage?

> Facet gives you a very simple way of visuallizing your data from multiple perspectives.

#### Can't we do without Facet?

> Of course we can, but to do that we have to **plot all plots individually** and they will have their own syntax. <br>
For ex. <br>
<br>
We have a Restaurant where people come and eat their food.
<br>
Two types of people - <u>Drunk</u> and <u>Non-Drunk</u> <br>
& <br>
Two types of time when they come - <u>Day</u> and <u>Night</u>
<br>
<br>
So, here I will have 4 possible combinations to choose from... <br>
 &nbsp; &nbsp;  &nbsp; &nbsp; `Drunk` &nbsp; comes at &nbsp; `Night` <br>
`Non-Drunk` &nbsp; comes at &nbsp; `Night` <br>
&nbsp; &nbsp;  &nbsp; &nbsp; `Drunk` &nbsp; comes at &nbsp; `Day` <br>
`Non-Drunk` &nbsp; comes at &nbsp; `Day` <br>

#### Clear advantage?

> Yes! Facet does it all for us, no matter there is third category of Gender, getting more combinations, the facet will allow us to do the same thing in just a single fucking line!

```python
Fecet = sns.FacetGrid(tips, row= 'sex', col= 'smoker')
```

### We've got Empty... waht? I am sorry, What?
**Explanation:** <br>
Here basically we told `sns` that,
    
    We: Hey!, I need the 'Combinations of the categories that I am giving you'
    sns: Sure, give me!
    
    We: Take 'Gender' as rows and 'Smoker' as columns
    sns: okay, let me find total distinct data ( categories )
    ...
    sns: Woh, I found 2 Genders and 2 Types of smokers, so you can have 4 plots.
         Give me WHAT do you need to see on those both parameters?
         and also give me the type of plot that you wanna plot!
         
    
    We: Alright!

```python
Fecet = sns.FacetGrid(tips, row= 'sex', col= 'smoker')
Fecet.map(sns.histplot, 'total_bill')
```

### What if I want to plot those kind of plots which requires TWO parametes, X and Y?
As of above, we only worked on the single parametered value, now multiple...

```python
Facet = sns.FacetGrid(tips, row= 'sex', col= 'smoker')
Facet.map(plt.scatter, 'total_bill', 'tip')
```
