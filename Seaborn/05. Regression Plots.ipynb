Before I write a code for LMplot, I want to clarify these things first.

<font color="#006bcf" face= "nexa"> **Why LM? I mean LM why?** </font>

    The actual thing is that, LM is the 'Advance' version of the simple plot which is called:
    'REGplot'.
    
    Yes, regplot() is the original one. 
    
<font color="#006bcf" face= "nexa"> **What is the difference?** </font>

    The main difference is, LMplot is like a Facetgrid. It gives you multiple comparisions while REGplot will
    give you only one plot at a time.
    
(And LM = Linear Model)
#### So, first we will explore the `regplot` and then will move for the `lmplot` where the actual lecture starts.

###  

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

###  Exploring `REGplot`

```python
sns.regplot(x= 'total_bill', y= 'tip', data= tips)
```

##### And yes, don't use` Labeled column`, as they won't work in regression!

```python
sns.regplot(x= 'total_bill', y= 'tip', data= tips, hue= 'sex')                        # See? hue <-- is not available in regplot
```

###  

Enough exploration in `regplot`!
### Exploring `LMplot`

```python
sns.lmplot(x= 'total_bill', y= 'tip', data= tips)
```

```python
sns.lmplot(x= 'total_bill', y= 'tip', data= tips, hue= 'sex')
```

#### Changing markers

```python
sns.lmplot(x= 'total_bill', y= 'tip', data= tips, hue= 'sex', markers= 's')
```

Worked, but let's seperate for the category...

```python
sns.lmplot(x= 'total_bill', y= 'tip', data= tips, hue= 'sex', markers= ['+','s'])
```

#### Buli Shuli! The data seems to dense!
What if I want to control the **size** of the markers?<br>
Actually if you see in the list of parameters, there is nothing called *marker_size*, insted you can find the *marker* as one of its parameters as they have given to access without writing the chaning code as it is so common to seperate the markers, but...

`I want to use the MATPLOTLIB parameters to control its look` what then? <br>
### Kindly use `scatter_kws` and pass the parameters as the dictionary!

```python
sns.lmplot(x= 'total_bill', y= 'tip', data= tips, hue= 'sex', scatter_kws= {'s': 200})
```

###  

###  Now, a real Power of LMplot...

```python
sns.lmplot(x= 'total_bill', y= 'tip', data= tips, hue= 'sex')            # BEFORE just in a single plot - Looked bad
```

```python
sns.lmplot(x= 'total_bill', y= 'tip', data= tips, col= 'sex', row= 'smoker')            # AFTER in multiple plots - Lookes good! ( Real Power )
```

##### Wait what? 
Yes! Just like Facetgrid from the Grid book! Row and Col parameters !

### Did you know?
'You also can add `hue` even after defining `row & col` parameteres'?

**Check it out**

```python
sns.lmplot(x= 'total_bill', y= 'tip', data= tips, row= 'day', col= 'smoker', hue= 'sex')
```

###  Amaizing!
It gives in a proper size that we can see... <br>
*But It still fills while screen.*

I want to control the size whether a Larger or Smaller...
Here is **HOW**

#### Well, with 

`aspect` & `size` parameters. <br>
(`size` has been renamed to `height`)

```python
sns.lmplot(x= 'total_bill', y= 'tip', data= tips, row= 'day', col= 'smoker', hue= 'sex',
          
          aspect= 0.5,
          height= 5)
```
