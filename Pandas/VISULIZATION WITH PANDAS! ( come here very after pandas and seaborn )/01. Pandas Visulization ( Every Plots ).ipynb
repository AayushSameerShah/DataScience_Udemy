___
**What .bar() actually does**

    It takes 'indices' of the df as 'x' and 'column' of df as 'y' ( Which can be multiple ) and basically plots them
    side-by-side or stacked manner.
    
    "It is used to compare the Columns on the given Row."
                                                    ∞ Aayush Shah

Side Note:
    
    You can basically use Multilevel index and plot the bar plot, but nothing will change as it will 
    consider each pair as the separate row! And you end up having an ugly lookig plot with densly 
    written lables!

Stacking Bar plot to make it look nicer!

```python
plt.style.use('dark_background') #Just to show case the Theme to be changed...
```

```python
df2.plot.bar(stacked = True)
```

**Scatter**

```python
df1.plot.scatter(x= 'A', y= 'B')
```

#### Adding 3rd Dimention
    Way 1: By Color         - c      (c = Color) 
    Way 2: By Size          - s      (s = Shape / Size)

I want to see the data by the 3rd dimention like column `'C'`

**Way 1: By Color**

```python
df1.plot.scatter(x= 'A', y= 'B', c= 'C')
```

**Way 2: By Size**

```python
df1.plot.scatter(x= 'A', y= 'B', s= df1['C'] * 30)
```

**Way 1 + Way 2 + Color!**

```python
df1.plot.scatter(x= 'A', y= 'B', s= df1['C'] * 30, c= 'C', cmap= 'seismic')           
```
