# Concatinating
" Gluing "

    __
    
    • This basically attatches two or more DFs together and make a new DF.
    • Here basically no condition is checked, just take A and B then make AB.
    __
    
See How...

```python
import pandas as pd
import numpy as np
```

```python
dataDict = {'A': [1,2,3,],
            'B': [4,5,6,],
            'C': [7,8,9,]}
```

```python
A_frame = pd.DataFrame(data= dataDict,
                       index= [1,2,3] )
```

```python
B_frame = pd.DataFrame(data= dataDict,
                       index= [4,5,6] )
```

```python
C_frame = pd.DataFrame(data= dataDict,
                       index= [7,8,9] )
```

```python
display(A_frame)
display(B_frame)
display(C_frame)
```

### Concatination

```python
pd.concat([A_frame, B_frame, C_frame])
```

#### Good, not some wild stuff...
Extra Col, Extra Row...

```python
dataDict = {'A': [1,2,3,111,],
            'B': [4,5,6,111,],
            'C': [7,8,9,111,]}

A_frame = pd.DataFrame(data= dataDict,
                       index= [1,2,3,4] )

B_frame = pd.DataFrame(data= dataDict,
                       index= [4,5,6,7] )

C_frame = pd.DataFrame(data= dataDict,
                       index= [7,8,9,0] )

pd.concat([A_frame, B_frame, C_frame])
```

#### That means, Extra row will work, and also it will 'Duplicate' the occurance.

```python
dataDict = {'A': [1,2,3,111,],
            'B': [4,5,6,111,],
            'C': [7,8,9,111,]}

_dataDict = {'A': [1,2,3,111,],
            'BB': [4,5,6,111,],         # For C_Frame ( Extra Column )
            'C': [7,8,9,111,]}

A_frame = pd.DataFrame(data= dataDict,
                       index= [1,2,3,4] )

B_frame = pd.DataFrame(data= dataDict,
                       index= [4,5,6,7] )

C_frame = pd.DataFrame(data= _dataDict,
                       index= [7,8,9,0] )

pd.concat([A_frame, B_frame, C_frame])
```

#### Aha! Extra column is added ! COOL

```python
dataDict = {'A': [1,2,3,111,],
            'B': [4,5,6,111,],
            'C': [7,8,9,111,]}

_dataDict = {'A': [1,2,3,111,],   # For C ( One Col less )
            'C': [7,8,9,111,]}

A_frame = pd.DataFrame(data= dataDict,
                       index= [1,2,3,4] )

B_frame = pd.DataFrame(data= dataDict,
                       index= [4,5,6,7] )

C_frame = pd.DataFrame(data= _dataDict,
                       index= [7,8,9,0] )

pd.concat([A_frame, B_frame, C_frame])
```

#### Once column less will work TOO!

## SO,

    > Columns can NOT be duplicate but Indexes can.
    > One more col or less won't matter.
    > Concatination is easy - Just see you are not making any mistake in creating them.
    
## But till here, it was on the axis = 0
#### Let's go for axis = 1 

```python
dataDict = {'A': [1,2,3,111,],
            'B': [4,5,6,111,],
            'C': [7,8,9,111,]}

_dataDict = {'A': [1,2,3,111,],
            'BB': [4,5,6,111,],
            'C': [7,8,9,111,]}

A_frame = pd.DataFrame(data= dataDict,
                       index= [1,2,3,4] )

B_frame = pd.DataFrame(data= dataDict,
                       index= [4,5,6,7] )

C_frame = pd.DataFrame(data= _dataDict,
                       index= [7,8,9,0] )

pd.concat([A_frame, B_frame, C_frame], axis=1)
```

###  Simple as that! No need of any Wild stuff

# Merging
' An SQL Join '

```python
AdataDict = {"A":[1,2,3],
            "B":[4,5,6],
            "Key":["K1","K2","K0"]}

BdataDict = {"AA":[1,2,3],
            "BB":[4,5,6],
            "Key":["K1","K2","K3"]}
A_frame = pd.DataFrame(data= AdataDict)
B_frame = pd.DataFrame(data= BdataDict)
```

```python
display(A_frame)
display(B_frame)
```

```python
pd.merge(A_frame,B_frame, how= "inner", on= "Key")
```

```python
pd.merge(A_frame,B_frame, how= "outer", on= "Key")
```

#### Key (on - parameter) must present in both DFs!

# Join
'An Indexial Merge'

```python
AdataDict = {"A":[1,2,3],
            "B":[4,5,6],
            }

BdataDict = {"AA":[11,22,33],
            "BB":[44,55,66],
            }
A_frame = pd.DataFrame(data= AdataDict)
B_frame = pd.DataFrame(data= BdataDict)
```

```python
display(A_frame)
display(B_frame)
```

```python
A_frame.join(B_frame)
```

#### LOOK, in join there is no - pd.join - there is left.join(right) 

```python
AdataDict = {"A":[1,2,3],
            "B":[4,5,6],
            }

BdataDict = {"AA":[11,22,33],
            "BB":[44,55,66],
            }
A_frame = pd.DataFrame(data= AdataDict, index= [1,2,3])
B_frame = pd.DataFrame(data= BdataDict, index= [2,3,4])
```

```python
A_frame.join(B_frame)
```
