## `axes.legend()` to print the legend
Error if no label found in any axes

````python
fig, axes = plt.subplots()
axes.plot(x,x**2, label="x Square")
axes.plot(x,x**3, label="x Qube")
axes.legend()
````

### In Case...

    If the legend overlaps the plot, you can specify the location for the legend to be located.

````python
fig, axes = plt.subplots()
axes.plot(x,x**2, label="x Square")
axes.plot(x,x**3, label="x Qube")
axes.legend(loc=10)
````
