# Controls over the lines to Show on the Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,10)
y = x ** 2
```

###  

```python
plt.plot(x,y)
```

###  

###  Now, let's see some of the attributes to be plotted on the plot!

#### color, alpha

```python
plt.plot(x,y, color= "red") #or use #FF22AA code
```

```python
plt.plot(x,y, color= "red")
plt.plot(y,x, color= "blue", alpha= 0.5)
```

###  

###  linewidth, linestyle

    linwidth = lw         BOTH  OF  THEM  
                          INTERCHANGEBLY
    linestyle = ls        WILL      WORK

```python
plt.plot(x,y, linewidth= 10)
```

```python
plt.plot(x,y, linewidth= 10, linestyle='--')
```

#### Other styles goes on...

```python
fig, axes = plt.subplots(2,2, figsize = (10,6))

axes[0][0].plot(x,y, lw= 10, ls= '--')
axes[0][1].plot(x,y, lw= 10, ls= '-.')
axes[1][0].plot(x,y, lw= 10, ls= ':')
axes[1][1].plot(x,y, lw= 10, ls= '-')


plt.tight_layout()
```

###  

##  P.hD on Marker.

     Category                  Controls
    ----------                ----------
      Shape                     marker
                                markersize
     
      Inner Shape               markerfacecolor

      Edge of Shape             markeredgewidth
                                markeredgecolor

### marker

```python
fig, axes = plt.subplots(5,2, figsize = (15,15))

axes[0][0].plot(x,y, marker= 'o', label= "o")
axes[0][1].plot(x,y, marker= '+', label= "+")
axes[1][0].plot(x,y, marker= ',', label= ",")
axes[1][1].plot(x,y, marker= '*', label= "*")
axes[2][0].plot(x,y, marker= 's', label= "s")
axes[2][1].plot(x,y, marker= '.', label= ".")
axes[3][0].plot(x,y, marker= '1', label= "1")
axes[3][1].plot(x,y, marker= '2', label= "2")
axes[4][0].plot(x,y, marker= '3', label= "3")
axes[4][1].plot(x,y, marker= '4', label= "4")

axes[0][0].legend()
axes[0][1].legend()
axes[1][0].legend()
axes[1][1].legend()
axes[2][0].legend()
axes[2][1].legend()
axes[3][0].legend()
axes[3][1].legend()
axes[4][0].legend()
axes[4][1].legend()

plt.tight_layout()
```

###  markersize

```python
fig, axes = plt.subplots(5,2, figsize = (15,15))

axes[0][0].plot(x,y, marker= 'o', markersize= 10, label= "o")
axes[0][1].plot(x,y, marker= '+', markersize= 10, label= "+")
axes[1][0].plot(x,y, marker= ',', markersize= 10, label= ",")
axes[1][1].plot(x,y, marker= '*', markersize= 10, label= "*")
axes[2][0].plot(x,y, marker= 's', markersize= 10, label= "s")
axes[2][1].plot(x,y, marker= '.', markersize= 10, label= ".")
axes[3][0].plot(x,y, marker= '1', markersize= 10, label= "1")
axes[3][1].plot(x,y, marker= '2', markersize= 10, label= "2")
axes[4][0].plot(x,y, marker= '3', markersize= 10, label= "3")
axes[4][1].plot(x,y, marker= '4', markersize= 10, label= "4")

axes[0][0].legend()
axes[0][1].legend()
axes[1][0].legend()
axes[1][1].legend()
axes[2][0].legend()
axes[2][1].legend()
axes[3][0].legend()
axes[3][1].legend()
axes[4][0].legend()
axes[4][1].legend()

plt.tight_layout()
```

###  markerfacecolor

```python
fig, axes = plt.subplots(5,2, figsize = (15,15))

axes[0][0].plot(x,y, marker= 'o', markersize= 10, markerfacecolor= "yellow", label= "o")
axes[0][1].plot(x,y, marker= '+', markersize= 10, markerfacecolor= "yellow", label= "+")
axes[1][0].plot(x,y, marker= ',', markersize= 10, markerfacecolor= "yellow", label= ",")
axes[1][1].plot(x,y, marker= '*', markersize= 10, markerfacecolor= "yellow", label= "*")
axes[2][0].plot(x,y, marker= 's', markersize= 10, markerfacecolor= "yellow", label= "s")
axes[2][1].plot(x,y, marker= '.', markersize= 10, markerfacecolor= "yellow", label= ".")
axes[3][0].plot(x,y, marker= '1', markersize= 10, markerfacecolor= "yellow", label= "1")
axes[3][1].plot(x,y, marker= '2', markersize= 10, markerfacecolor= "yellow", label= "2")
axes[4][0].plot(x,y, marker= '3', markersize= 10, markerfacecolor= "yellow", label= "3")
axes[4][1].plot(x,y, marker= '4', markersize= 10, markerfacecolor= "yellow", label= "4")

axes[0][0].legend()
axes[0][1].legend()
axes[1][0].legend()
axes[1][1].legend()
axes[2][0].legend()
axes[2][1].legend()
axes[3][0].legend()
axes[3][1].legend()
axes[4][0].legend()
axes[4][1].legend()

plt.tight_layout()
```

###  markeredgecolor

```python
fig, axes = plt.subplots(5,2, figsize = (15,15))

axes[0][0].plot(x,y, marker= 'o', markersize= 10, markerfacecolor= "yellow", markeredgewidth= 5, label= "o")
axes[0][1].plot(x,y, marker= '+', markersize= 10, markerfacecolor= "yellow", markeredgewidth= 5, label= "+")
axes[1][0].plot(x,y, marker= ',', markersize= 10, markerfacecolor= "yellow", markeredgewidth= 5, label= ",")
axes[1][1].plot(x,y, marker= '*', markersize= 10, markerfacecolor= "yellow", markeredgewidth= 5, label= "*")
axes[2][0].plot(x,y, marker= 's', markersize= 10, markerfacecolor= "yellow", markeredgewidth= 5, label= "s")
axes[2][1].plot(x,y, marker= '.', markersize= 10, markerfacecolor= "yellow", markeredgewidth= 5, label= ".")
axes[3][0].plot(x,y, marker= '1', markersize= 10, markerfacecolor= "yellow", markeredgewidth= 5, label= "1")
axes[3][1].plot(x,y, marker= '2', markersize= 10, markerfacecolor= "yellow", markeredgewidth= 5, label= "2")
axes[4][0].plot(x,y, marker= '3', markersize= 10, markerfacecolor= "yellow", markeredgewidth= 5, label= "3")
axes[4][1].plot(x,y, marker= '4', markersize= 10, markerfacecolor= "yellow", markeredgewidth= 5, label= "4")

axes[0][0].legend()
axes[0][1].legend()
axes[1][0].legend()
axes[1][1].legend()
axes[2][0].legend()
axes[2][1].legend()
axes[3][0].legend()
axes[3][1].legend()
axes[4][0].legend()
axes[4][1].legend()

plt.tight_layout()
```

###  markeredgecolor

```python
fig, axes = plt.subplots(5,2, figsize = (15,15))

axes[0][0].plot(x,y, marker= 'o', markersize= 10, markerfacecolor= "yellow", markeredgecolor= "g", label= "o")
axes[0][1].plot(x,y, marker= '+', markersize= 10, markerfacecolor= "yellow", markeredgecolor= "g", label= "+")
axes[1][0].plot(x,y, marker= ',', markersize= 10, markerfacecolor= "yellow", markeredgecolor= "g", label= ",")
axes[1][1].plot(x,y, marker= '*', markersize= 10, markerfacecolor= "yellow", markeredgecolor= "g", label= "*")
axes[2][0].plot(x,y, marker= 's', markersize= 10, markerfacecolor= "yellow", markeredgecolor= "g", label= "s")
axes[2][1].plot(x,y, marker= '.', markersize= 10, markerfacecolor= "yellow", markeredgecolor= "g", label= ".")
axes[3][0].plot(x,y, marker= '1', markersize= 10, markerfacecolor= "yellow", markeredgecolor= "g", label= "1")
axes[3][1].plot(x,y, marker= '2', markersize= 10, markerfacecolor= "yellow", markeredgecolor= "g", label= "2")
axes[4][0].plot(x,y, marker= '3', markersize= 10, markerfacecolor= "yellow", markeredgecolor= "g", label= "3")
axes[4][1].plot(x,y, marker= '4', markersize= 10, markerfacecolor= "yellow", markeredgecolor= "g", label= "4")

axes[0][0].legend()
axes[0][1].legend()
axes[1][0].legend()
axes[1][1].legend()
axes[2][0].legend()
axes[2][1].legend()
axes[3][0].legend()
axes[3][1].legend()
axes[4][0].legend()
axes[4][1].legend()

plt.tight_layout()
```
