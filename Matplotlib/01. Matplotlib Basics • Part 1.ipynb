##### Basic Intro

    Here, the term 'Basic' means the simple and initial way to approach the multi plotting technique and for
    another technique I would encourage you to scroll down and head to the 'Advanced Approach'.
    
    For now, Intro of this:
        The subplot basically, well as name suggests, 'subplots' the original plot.
        So, if the original plot is 10 * 10 larger, lets say, then putting two charts in the single row 
        means 5 * 5 per chart and put them side by side.
        
        If you put 2 charts but in a chart per row, then 5 * 5 will be assigned but the charts will be put
        in the top to bottom manner.
        
        And there, you can specify any ( I guess any ) amount of charts in that given plot. 
        [ 4 charts 2 per row and 2 per col will end up providing 2.5 * 2.5 area each,
          you should try some wild stuff... ]
          
             >>>   plt.subplot(row, column, onWhich)
             >>>   plt.subplot(1  ,      2,       1) 
             
             (No need to explain, you are Intelligent!)

````python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y,'b')
````

### Advance Approach
    
    Also called the "Object Oriented Approach"
    Steps:
    
    1) Define a Figure - 'fig' obj 
    2) Set parameters of that 'fig' i.e. Axis
    3) Plot with that 'fig'
   
See them in action !

````python
fig = plt.figure()
````

<font color="green"> Question? </font> <br>
Why ```fig = plt.figure()``` returned a string of obj location even if we are assigning it to the variable?
<br> <br>
<font color="green"> Answer: </font> <br>
Because they are printing that line, not returning it.

````python
#Repeating again,to make sense

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #For larger, use [1,1,2,2]
````

### Note

    [0.1,    0.1,   0.8,    0.8]
    Left, Bottom, Width, Height]
