#### Now what?
Now we are going to impute the data in.  
Well, that can be done through ML in which we would take all features into an account, but here HE is not going to go with that. ( I guess we could do that! )  

In short, we will fill in the values by mean of each classes.

```python
train.Age.isnull().sum()
```

```python
means = train.groupby('Pclass')['Age'].mean()
```

```python
means
```

```python
def fill(class_age):
    if pd.isnull(class_age[1]):
        return means[class_age[0]]
    return class_age[1]
```

```python
means[0:]
```

```python
train['Age'] = train[['Pclass', 'Age']].apply(fill, axis=1)
```

### *Fixing*...

```python
test
```

```python
means = test.groupby('Pclass')['Age'].mean()
```

```python
means
```

```python
def fill_age(class_age):
    if pd.isnull(class_age[1]):
        return means[class_age[0]]
    return class_age[1]
```

```python
test['Age'] = test[['Pclass', 'Age']].apply(fill_age, axis= 1)
```
